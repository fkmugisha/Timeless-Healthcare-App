```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Elbow method to determine the number of K in Kmeans Clustering
coords = df[['X','Y']]
coords.head()
```

```python
distortions = []
K = range(1,20)
for k in K:
    kmeansModel = KMeans(n_clusters=k)
    kmeansModel = kmeansModel.fit(coords)
    distortions.append(kmeansModel.inertia_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
plt.plot(K, distortions, marker='o')
plt.xlabel('k')
plt.ylabel('Distortions')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++')
kmeans.fit(coords)
y = kmeans.labels_
```

```python
df['cluster'] = kmeans.predict(df[['X','Y']])
df.head()
```

```python
def recommend_health_centers(df, longitude, latitude, emergency="", operator=""):
    # Predict the cluster for longitude and latitude provided
    cluster = kmeans.predict(np.array([longitude,latitude]).reshape(1,-1))[0]
    print(cluster)
    
    if operator == "private" and emergency =="":
        df = df[df['operator'].str.lower()=='private']
    elif operator == "government" and emergency=="":
        df = df[df['operator'].str.lower()!='private']
    elif operator == "" and emergency =="yes":
        df = df[df['emergency'].str.lower()=='yes']
    elif operator == "private" and emergency == "yes":
        df = df[df['operator'].str.lower()=='private']
        df = df[df['emergency'].str.lower()=='yes']
    elif operator == "private" and emergency == "no":
        df = df[df['operator'].str.lower()=='private']
    elif operator == "government" and emergency =="yes":
        df = df[df['operator'].str.lower()!='private']
        df = df[df['emergency'].str.lower()=='yes']
    elif operator == "government" and emergency =="no":
        df = df[df['operator'].str.lower()!='private']
    else:
        df = df
        
   
    # Get the best hospitals according to users inputs in this cluster
    return  df[df['cluster']==cluster].iloc[0:5][['Name', 'X','Y', 'operator', 'emergency']]
```

```python
recommend_health_centers(df, longitude='32.745920', latitude='0.360830',emergency='yes', operator='private' )
```
